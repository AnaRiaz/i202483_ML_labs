```python
# P(yes) , P(No)
Y = 0 
N = 0 
L = len(df['play'])
for i in range(0,len(df["play"])):
    if df['play'][i] =='Yes':
        Y+=1
    else:
        N += 1
print(Y,N)
P_Yes = Y/len(df['play'])
P_No = N/len(df['play'])
```

###### class entropy H(D):

###### Entropy of column 'wind' wrt target column

```python
#feature: 'wind'
weak = 0
strong = 0
Wk_No =0
Wk_Yes=0
St_No =0
St_Yes=0
for i in range(0,len(df['wind'])):
    if df['wind'][i]=="Weak":
        weak +=1
    if df['wind'][i]=="Strong":
        strong += 1
for i in range(0,len(df["wind"])):
    if df['wind'][i]=="Weak" and df['play'][i]=="Yes":
        Wk_Yes += 1
    if df['wind'][i]=="Weak" and df['play'][i]=="No":
        Wk_No += 1
    if df['wind'][i]=='Strong' and df['play'][i]=="Yes":
        St_Yes +=1
    if df['wind'][i]=='Strong' and df['play'][i]=='No':
        St_No += 1
Prob_Wk_Yes = Wk_Yes / weak 
Prob_Wk_No = Wk_No / weak
Prob_St_Yes = St_Yes / strong
Prob_St_No = St_No / strong
print("P(Yes|Weak): ", Prob_Wk_Yes, "\nP(No|Weak): " ,Prob_Wk_No , "\nP(Yes|Strong): ",Prob_St_Yes, "\nP(No|Strong):",Prob_St_No )
```

```python
# entropy of weak:
E_weak = -1*((Prob_Wk_Yes* (np.log2(Prob_Wk_Yes))) + Prob_Wk_No* (np.log2(Prob_Wk_No)))
# entropy of strong:
E_strong = -1*((Prob_St_Yes* (np.log2(Prob_St_Yes))) + Prob_St_No* (np.log2(Prob_St_No)))
print("Entropy of week: " , E_weak ,"\nEntropy of strong: ",E_strong)
```

```python
high = 0
norm = 0
H_Yes = 0
H_No = 0
Norm_No = 0
Norm_Yes = 0
for i in range(0,len(df['humidity'])):
    if df['humidity'][i]=="High":
        high +=1
    if df['humidity'][i]=="Normal":
        norm += 1
for i in range(0,len(df["humidity"])):
    if df['humidity'][i]=="High" and df['play'][i]=="Yes":
        H_Yes += 1
    if df['humidity'][i]=="High" and df['play'][i]=="No":
        H_No += 1
    if df['humidity'][i]=='Normal' and df['play'][i]=="Yes":
        Norm_Yes +=1
    if df['humidity'][i]=='Normal' and df['play'][i]=='No':
        Norm_No += 1
```

```python
prob_H_Yes = H_Yes/high
prob_H_No = H_No = H_No/high
prob_Norm_Yes = Norm_Yes/norm
prob_Norm_No = Norm_No/norm
print(prob_Norm_Yes)
```

```python
E_High = -1 * ((prob_H_Yes*(np.log2(prob_H_Yes))) + prob_H_No*(np.log2(prob_H_No)))
print("Entropy of high: " ,E_High)
E_Normal = -1 * ((prob_Norm_Yes*(np.log2(prob_Norm_Yes))) + prob_Norm_No*(np.log2(prob_Norm_No)))
print("Entropy of normal: " , E_Normal)
```

###### Entropy of column 'temp' wrt target column

```python
# temp
hot = 0
mild = 0
cool = 0
hot_Yes = 0
mild_Yes = 0
cool_Yes = 0
hot_No = 0
mild_No = 0
cool_No = 0
for i in range(0,len(df['temp'])):
    if df['temp'][i]=="Hot":
        hot +=1
    if df['temp'][i]=="Mild":
        mild += 1
    if df['temp'][i]=="Cool":
        cool += 1
for i in range(0,len(df["temp"])):
    if df['temp'][i]=="Hot" and df['play'][i]=="Yes":
        hot_Yes += 1
    if df['temp'][i]=="Hot" and df['play'][i]=="No":
        hot_No += 1
    if df['temp'][i]=='Mild' and df['play'][i]=="Yes":
        mild_Yes +=1
    if df['temp'][i]=='Mild' and df['play'][i]=='No':
        mild_No += 1
    if df['temp'][i]=='Cool' and df['play'][i]=="Yes":
        cool_Yes +=1
    if df['temp'][i]=='Cool' and df['play'][i]=='No':
        cool_No += 1
      
```

```python
prob_hot_No = hot_No/hot
prob_hot_yes = hot_Yes/hot
prob_mild_yes = mild_Yes/mild
prob_mild_No = mild_No/mild
prob_cool_yes = cool_Yes/cool
prob_cool_No  = cool_No/cool
```

```python
E_hot  = -1* ( (prob_hot_No*(np.log2(prob_hot_No)))  +  (prob_hot_yes*(np.log2(prob_hot_yes))) )
E_mild = -1* ( (prob_mild_yes * (np.log2(prob_mild_yes ))) + (prob_mild_No *(np.log2(prob_mild_No ))) )
E_cool = -1* ( (prob_cool_yes)*(np.log2(prob_cool_yes)) + (prob_cool_No*(np.log2(prob_cool_No))) )
```

```python
# for clas entropy
def classEntropy(col):
    unq,count = np.unique(col , return_counts=True)
#    print(unq,count)
    s = 0    
    for i in count:
        s += i/(len(col)) * np.log2(i/(len(col)))    
    s *= (-1)
    return s
```

```python
def informationGain (v1,v2):   
    unq_v1 , countV1 = np.unique(v1 , return_counts=True)
    unq_v2 , countV2 = np.unique(v2 , return_counts=True)
#   print(unq_v2 , countV2)
    arr_yes = []
    arr_no  = []    
    for j in unq_v1: 
        yeses = 0
        nos  = 0         
        for k in range(len(v1)):
            if j == v1[k]:
                if  v2[k] == 'Yes':
                    yeses += 1
                else:
                    nos += 1
        arr_yes.append (yeses)
        arr_no.append (nos)        
    summ = 0
    arr_sum = []
    for i in range(0,len(unq_v1)):
        summ = 0        
        if (arr_yes[i]== 0):
            summ += -1 *((arr_no[i]/(countV1[i]) * np.log2(arr_no[i]/(countV1[i]))) )
            
        elif (arr_no[i]== 0):
            summ += -1 *((arr_yes[i]/(countV1[i]) * np.log2(arr_yes[i]/(countV1[i]))) )         
        else:
            summ += -1 *((arr_yes[i]/(countV1[i]) * np.log2(arr_yes[i]/(countV1[i]))) )
            summ += -1 *((arr_no[i]/(countV1[i]) *np.log2(arr_no[i]/(countV1[i]))) )      
        arr_sum.append(summ)  
#     print(arr_sum)
    E_split = 0
    
    for i in range(len(countV1)):
        E_split += arr_sum[i] * countV1[i]/len(v2)  
#     print(E_split)
    Ent_Class = classEntropy(v2)
    I = Ent_Class - E_split  #EntClass = E_Class -> calculated in cell 3    
    print((I))    
    return I
```
