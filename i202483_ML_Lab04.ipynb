### Task 08

```python
cities = ['A','B','C','D','E']
person1 = ""
sample = []
for i in cities:
    person1 = ""
    person1 += i
    person3 = person1
    for j in cities:
        person1 = person3
        person1 += j
        person2 = person1
        for k in cities:
            person1 = person2
            person1 += k
            sample.append(person1)
sample
```

```python
event = []
for i in sample:
    idx1 = ""
    idx2 = ""
    idx3 = ""
    idx1 = i[0]
    idx2 = i[1]
    idx3 = i[2]
    if (idx1==idx2) or (idx2==idx3) or (idx1==idx3):
        event.append(i)
occurance = len(event)
prob = occurance/len(sample)*100
print("Probablity of the event:",prob)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
xAxis = fig.add_axes([0,0,1,1])
xAxis.set_ylabel('Probability')
xAxis.set_title('Probability Distribution function')
title = ['Same city', 'Not the same city']
Probablities = [52,48]
xAxis.bar(title,Probablities)
plt.show()
```

```python
df = pd.read_csv("Housing.csv")
```

```python
df
```

```python
import statistics as st
from scipy import stats
beds = df ["bedrooms"]
mean = np.mean(beds, axis = 0)
print("Mean of one feature: ", mean)
std_dev = beds.std()
print("Standard Dev: ",std_dev)
variance = std_dev**2
print("Variance: ",variance)
zScore = abs(stats.zscore(df['bedrooms']) )
print("Z Score: \n" ,zScore)
```

### Task 02

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report 
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

column = ['mainroad', 'guestroom', 'basement', 'hotwaterheating','airconditioning', 
          'prefarea', 'furnishingstatus']
df[column] = df[column].apply(LabelEncoder().fit_transform)
label = df["furnishingstatus"]
x = df.drop("furnishingstatus", axis=1)
Xtrain, Xtest, ytrain, ytest = train_test_split(x, label , test_size = 0.2 , random_state = 0)

clf = RandomForestClassifier()
clf.fit(Xtrain, ytrain)
p = clf.predict(Xtest)
print ("Accuracy : " , accuracy_score(ytest,p)*100)  
print("Report : \n", classification_report(ytest, p))
print("F1 Score : ",f1_score(ytest, p, average='macro')*100)
```

### Task 03

```python
arr = np.array(df.drop(["furnishingstatus"], axis=1))
means = np.mean(arr , axis=0)
o = np.ones(arr.shape)
print(o.shape)
print(means.shape)
mat = means*o # matrix with the copies of mean values
Z = arr-mat
SD = np.std(arr,axis=0)        #standard dev
z = Z/SD
```

```python
print(z)
```

```python
# z represents the normalized dataset
x = z
```

```python
print(x.shape)
```

```python
#converting the nrmalized array into df
newdf = pd.DataFrame(x,columns=['price','area','length','width','bedrooms','bathrooms',
           'stories','mainroad','guestroom','basement','hotwaterheating',
            'airconditioning','parking','prefarea' ])
```
