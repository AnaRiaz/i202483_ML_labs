# Task 1

```python
import numpy as np
import pandas as pd
import random
```

```python
df= pd.read_csv("Advertising.csv")
df
```

```python
# putting the "sales" coulmn in Y
Y = df.iloc[:,-1:]
Y = np.array(Y)
print(Y)
```

```python
# selecting all the columns except first two in X
X = df.iloc[: , 1:-1]
# adding a column of ones in the training dataset with dim= dx1

X['X_not'] = np.ones (200)
X = np.array(X)
print(X)
```

### Initializing Alpha and Thetas

```python
# Since number of thetas = number of features + 1 
# but the addition of a "ones" column in the datset makes it -> number of thetas = number of features
alpha = 0.0001
thetas = np.array([[80] , [50] , [65] , [1]])
```

### Defining  Functions 

```python
def hypothesis(X1,thetas1):
    return np.dot(X1,thetas1)
```

```python
def loss_fun(y_act,y_pred):
    return np.mean((y_act-y_pred)**2)
```

```python
def gradient_descent(alpha,thetas2,g_thetas):
    return (thetas2-alpha*g_thetas)
```

```python
def der_loss_fun(x,y,thetas3):
    hp=hypothesis(x,thetas3)
    diff = (y-hp).T
    f1=np.dot(diff,X)
    f1=-2*f1/float(x.shape[0])
    return f1
```

```python
### checking if all our made arrays have consistent shapes
print(X.shape)
print(Y.shape)
print(thetas.shape)
```

### caluculating predicted Y

```python
y_pred = hypothesis(X,thetas)
y_pred.shape
```

```python
Y.shape
```

### Finding Mean Squared Error

```python
mse = loss_fun(Y,y_pred)
mse
```

### derivative of loss function

```python
gradient_thetas = der_loss_fun(X,Y,thetas)
gradient_thetas = gradient_thetas.T
gradient_thetas.shape
```

### updating thetas

```python
new_thetas =gradient_descent(0.001,thetas,gradient_thetas)
```

```python
new_thetas
```

```python
h1 = hypothesis(X,new_thetas)
```

```python
loss_fun(Y,h1)
```

```python
# gradient descent for 100 iterrations
for i in range(100):
    y_pred = hypothesis(X , new_thetas)
    gradient_thetas = der_loss_fun(X , Y , new_thetas)
    gradient_thetas = gradient_thetas.T
    new_thetas = gradient_descent (alpha , new_thetas , gradient_thetas)
    print(loss_fun(Y , y_pred))
```

```python
# final thetas
gradient_thetas
```

```python
thetas
```

The original thetas and final thetas tell the linear relationship between features and outputs
