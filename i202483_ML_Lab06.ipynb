```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("play_tennis.csv")
df.head(14)
```

```python
df.drop(['day'],axis=1,inplace = True)
df
```

```python
# col1 -> outlook
# col2 -> temp
# col3 -> humidity
# col4 -> wind
def NaiveBayes (feature1,feature2,feature3,feature4):
    col1_no = 0 
    col1_yes = 0 
    col2_no = 0
    col2_yes = 0
    col3_no = 0
    col3_yes = 0
    col4_no = 0
    col4_yes = 0
    No = 0 
    Yes = 0
    
    # when the outlook value is equal to the feature1 passed in the function 
    # and the label class is 'no'
    for i in range (0,len (df['outlook'])):
        if (df['outlook'][i] == feature1) and (df['play'][i]=='No'):
            col1_no += 1

    # when the temp value is equal to the feature2 passed in the function
    # and the label class is 'no'            
    for i in range (0,len(df['temp'])):    
        if (df['temp'][i] == feature2) and (df['play'][i]=='No'):
            col2_no += 1
            
    # when the humidity value is equal to the feature3 passed in the function 
    # and the label class is 'no'            
    for i in range (0,len(df['humidity'])):
        if (df['humidity'][i] == feature3) and (df['play'][i]=='No'):
            col3_no += 1
            
    # when the wind value is equal to the feature4 passed in the function  
    # and the label class is 'no'            
    for i in range (0,len(df['wind'])):
        if (df['wind'][i] == feature4) and (df['play'][i]=='No'):
            col4_no += 1

    # when the outlook value is equal to the feature1 passed in the function  
    # and the label class is 'Yes'            
    for i in range (0,len(df['outlook'])):
        if (df['outlook'][i] == feature1) and (df['play'][i]=='Yes'):
            col1_yes += 1
            
    # when the temp value is equal to the feature1 passed in the function  
    # and the label class is 'Yes'
    for i in range (0,len(df['temp'])):    
        if (df['temp'][i] == feature2) and (df['play'][i]=='Yes'):
            col2_yes += 1
            
    # when the humidity value is equal to the feature1 passed in the function  
    # and the label class is 'Yes'
    for i in range (0,len(df['humidity'])):
        if (df['humidity'][i] == feature3) and (df['play'][i]=='Yes'):
            col3_yes += 1
            
    # when the wind value is equal to the feature1 passed in the function  
    # and the label class is 'Yes'
    for i in range (0,len(df['wind'])):
        if (df['wind'][i] == feature4) and (df['play'][i]=='Yes'):
            col4_yes += 1
            
    for i in range (0, len(df['play'])):
        if (df['play'][i]== 'No'):
            No += 1
        if (df['play'][i] == 'Yes'):
            Yes += 1
    # formula : product probs of independent features       
    Prob_No = (col1_no/No)*(col2_no/No)*(col3_no/No)*(col4_no/No)*(No/len(df['outlook']))
    Prob_Yes = (col1_yes/Yes)*(col2_yes/Yes)*(col3_yes/Yes)*(col4_yes/Yes)*(Yes/len(df['outlook']))
    
    print ("\nP(Not Playing) = " , Prob_No*100)
    print ("P(Playing) = ", Prob_Yes*100)
    
    if (Prob_Yes > Prob_No):
        print ("The Players will play\n")
    else :
        print ("The PLayers will not play")
```

```python
v1 = input ("Weather Outlook:")
v2 = input ("Temperature:")
v3 = input ("Humitidy level:")
v4 = input ("Wind condition:")
NaiveBayes(v1,v2,v3,v4)
```

```python
Positive = "I love Mangoes mangoes are a lot tastier than other fruits I like mangoes because they are good Apples are good too but I prefer mangoes People say mango is the king of fruit so that is why mangoes are good mangoes are good because they are juicy"
Negative = "Eating too many Mangoes can have a negative impact on your health mangoes are good but not too good, I like apples more than mangoes I don’t like mangoes because they are expensive Mangoes are not good I don’t like mangoes"
```

```python
Positive = Positive.lower()
Negative = Negative.lower()
```

##### Naive Bayes

```python
def NaiveBayesClf(sentence):
    sentence = sentence.lower()
    test = sentence.split() # puts every word in the sentence on each index of the array
    Pos = Positive.split() # Neg array mein negative sentence ka every word para hai
    word1 = 0 # times the word one of the test array was found in the positive sentence
    word2 = 0 # times the word two of the test array was found in the positive sentence
    word3 = 0 # times the word three of the test array was found in the positive sentence
    for i in range(0,len(Pos)):
        if (Pos[i] == test[0]):
            word1 += 1         
        if (Pos[i] == test[1]):
            word2 += 1   
        if (Pos[i] == test[2]):
            word3 += 1           
    L = len(Pos)
    #Probability that sentence is positive:
    Prob_Positive = ((word1/L)*(word2/L)*(word3/L)*(5/11))
    Neg = Negative.split()   #splitting the negative sentence
    w1 = 0   # times the word one of the test array was found in the negative sentence
    w2 = 0   # times the word one of the test array was found in the negative sentence
    w3 = 0   # times the word one of the test array was found in the negative sentence
    for i in range (0, len(Neg)):
        if (Neg[i] == test[0]):
            w1 += 1
        if (Neg[i] == test[1]):
            w2 += 1
        if (Neg[i] == test[2]):
            w3 += 1        
    #negative probability 
    Prob_Negative = ((w1/len(Neg))*(w2/len(Neg))*(w3/len(Neg))*(6/11))    
    if Prob_Positive > Prob_Negative:
        print("Class: Positive")
    else:
        print("Class: Negative")        
```
