```python
df1 = df[df["-1.0000000e+00"]==1.0]
```

```python
train_size = 17 # 17(train):7(test)
test_size = 7
```

```python
ecg1 = pd.concat([df1[0:train_size],df2[0:test_size]])
#[0:17],[0:7]
ecg2 = pd.concat([df1[train_size:2*train_size],df2[test_size:2*test_size]])
#[17:34],[7;14]
ecg3 = pd.concat([df1[2*train_size:3*train_size],df2[2*test_size:3*test_size]])
#[34:51],[14:21
ecg4 = pd.concat([df1[3*train_size:70],df2[3*test_size:31]])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report 
from sklearn.ensemble import BaggingClassifier
from sklearn.utils import shuffle
from sklearn.model_selection import KFold
```

```python
# shuffeling all the dataframes to avoid any patterned results
data1 = shuffle(ecg1)
data2 = shuffle(ecg2)
data3 = shuffle(ecg3)
data4 = shuffle(ecg4)
```

```python
manual_acc=[]
sklearn_acc=[]
```

```python
#using data4 for testing
X1 = pd.concat([data1, data2, data3], axis=0)
Y1 = data4
xtrain = X1.drop('-1.0000000e+00',axis=1)
ytrain = X1["-1.0000000e+00"]
xtest = Y1.drop('-1.0000000e+00',axis=1)
ytest = Y1["-1.0000000e+00"]

clf = BaggingClassifier(KNeighborsClassifier(),max_samples=0.5, max_features=0.5)
clf.fit(xtrain, ytrain)
prediction = clf.predict(xtest)
a=accuracy_score(ytest,prediction)*100
manual_acc.append(a)
print ("Accuracy :\n " , a)
print("Report : \n\n", classification_report(ytest, prediction))
print("F1 Score : ",f1_score(ytest, prediction, average='macro')*100)
```

```python
#sklearn

X = df.drop('-1.0000000e+00',axis=1)
Y = df['-1.0000000e+00']
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.3, random_state = 27)
clf = BaggingClassifier(KNeighborsClassifier(),max_samples=0.5, max_features=0.5)
clf.fit(xtrain, ytrain)
prediction = clf.predict(xtest)
print ("Accuracy :\n " , accuracy_score(ytest,prediction)*100)  
print("Report : \n\n", classification_report(ytest, prediction))
print("F1 Score : ",f1_score(ytest, prediction, average='macro')*100)
```

```python
kfold = KFold(n_splits = 4,shuffle = True)
sklearn_acc=[]
print("Accuracies with sklearn:" )
for i, j in kfold.split(X): 
    xtrain = X.iloc[i]
    xtest  = X.iloc[j]
    ytrain = Y.iloc[i]
    ytest  = Y.iloc[j]    
    clf2 = BaggingClassifier(KNeighborsClassifier(),max_samples=0.5, max_features=0.5)
    clf2.fit(xtrain, ytrain)
    pred = clf2.predict(xtest)
    clf2.score(xtest,ytest)
    
    print (accuracy_score(ytest,pred)*100)
    sklearn_acc.append(accuracy_score(ytest,pred)*100)
```

```python
print("Without Sklearn:",manual_acc)
print("With Sklearn:",sklearn_acc)
```

```python
#checking which method gives better accuracies
sum1= sum(manual_acc)
sum2= sum(sklearn_acc)

Average1 = sum1//4
Average2 = sum2//4
print("Average Acuuracy using numpy:",Average1)
print("Average Acuuracy using skLearn:",Average2)

```
